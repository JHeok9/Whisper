```python
# whisper 불러와서 모델 정의
import whisper

model = whisper.load_model('base')
```

```python
# 샘플용 한국어 음성데이터 https://livros.tistory.com/209
from IPython.display import Audio
Audio('./data/sample.mp3')
```

```python
# 음성데이터 텍스트로 변환 transcribe
# 결과값은 text, segments, language 세개의 key값을 가진 딕셔너리형태
result = model.transcribe('./data/sample.mp3')
result
```

```python
# 변환된 텍스트만 출력하기
result['text']
```

```python
# segments는 리스트와 딕셔너리형태
# 한 문장의 시작과 끝을 알수있음
print('시작시간:',result['segments'][0]['start'])
print('끝시간:',result['segments'][0]['end'])
print('text:',result['segments'][0]['text'])
```

```python
# for문 활용하여 분석된 음성파일의 문장별 시작시간, 끝시간, 텍스트 출력
for i in result['segments']:
    print(f'[ {i["start"]} -> {i["end"]} ] {i["text"]} ')
```

```python
# 음성파일 변환시 영문으로 변역하여 변환
result1 = model.transcribe('./data/sample.mp3', task = 'translate')
result1['text']
```
